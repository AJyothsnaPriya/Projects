```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_boston
boston=load_boston()
boston
```

```python
#Transform the data in to dataframes
#data input variable
#feature names-column names
#target=output
```

```python
df_x=pd.DataFrame(boston.data,columns=boston.feature_names)
df_y=pd.DataFrame(boston.target)
```

```python
#Get some statics from the dataset
df_x.describe()
```

```python
#creating linear model
reg=LinearRegression()
```

```python
#Split the data into 67% training and 33% testing
x_train,x_test,y_train,y_test=train_test_split(df_x,df_y,test_size=0.33,random_state=0)
```

```python
#Training the data
reg.fit(x_train,y_train)
```

```python
#print the coefficients/weights for each feature/column
reg.coef_
```

```python
#print the predictions on our test data
y_pred=reg.predict(x_test)
y_pred
```

```python
#print acutual value
print(y_test)
```

```python
#check the model performace/accuracy using MSE-numpy
print(np.mean(y_pred-y_test)**2)
```

```python
#check the model performace/accuracy using MSE-numpy-sklearn metrics

from sklearn import metrics
from sklearn.metrics import mean_squared_error
print( mean_squared_error(y_test,y_pred))
```
